## Özet
Bu rapor, Luvionis platformuna entegre edilmiş örnek bir online satış veri setinin
nedensel analizini içermektedir. Amacımız; ülke bazlı satış performansları, ürün
kategori dağılımları ve müşteri harcama örüntüleri arasındaki ilişkileri ortaya
koyarak e-ticaret yöneticilerine optimize edilmiş kararlar sunmaktır.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.io as pio

pio.renderers.default = "notebook"
sns.set_style("whitegrid")
plt.rcParams["figure.figsize"] = (10, 5)

df = pd.read_csv("data/online_retail_II.csv", encoding="utf-8", on_bad_lines="skip")

# Veri temizliği
df = df.dropna(subset=["Customer ID"])         # Müşteri ID'si olmayanları çıkar
df = df[df["Quantity"] > 0]                    # İptal edilen siparişleri çıkar
df = df[df["Price"] > 0]                       # Geçersiz fiyatları çıkar
df["TotalPrice"] = df["Quantity"] * df["Price"] # Satır bazı ciro
df["InvoiceDate"] = pd.to_datetime(df["InvoiceDate"])
df["Month"] = df["InvoiceDate"].dt.month
df["DayOfWeek"] = df["InvoiceDate"].dt.dayofweek  # 0=Pazartesi, 6=Pazar

df.head()
```

## 1. Veri Seti Hakkında
Bu veri seti, bir UK merkezli online perakende firmasının 2009-2011 yılları 
arasındaki işlemlerini içermektedir. Luvionis analitik motoruna göre 
ön işleme sonrası kullanılan kayıt sayısı aşağıda görülmektedir.

```python
print("Toplam işlem sayısı:", len(df))
print("Benzersiz müşteri sayısı:", df["Customer ID"].nunique())
print("Kapsanan ülke sayısı:", df["Country"].nunique())
print("\nEksik değerler:\n", df.isnull().sum())
```

## 2. Tanımsal İstatistikler

```python
# İstatistik 1: Genel özet
display(df[["Quantity", "Price", "TotalPrice"]].describe().round(2))

# İstatistik 2: Ülkeye göre toplam ciro (UK hariç — daha okunabilir)
display(df[df["Country"] != "United Kingdom"].groupby("Country")["TotalPrice"].sum().sort_values(ascending=False).head(10))

# İstatistik 3: Korelasyon matrisi
display(df[["Quantity", "Price", "TotalPrice"]].corr().round(3))

# İstatistik 4: Aya göre ortalama ve toplam ciro
display(df.groupby("Month")["TotalPrice"].agg(["mean", "sum", "count"]).round(2))
```

## 3. Görsel Analiz — Luvionis Dashboard Prototipleri

```python
fig, ax = plt.subplots()
df["TotalPrice"].clip(upper=500).hist(bins=50, ax=ax, color="steelblue", edgecolor="white")
ax.set_title("Sipariş Tutarı Dağılımı (0-500£ arası)")
ax.set_xlabel("Sipariş Tutarı (£)")
ax.set_ylabel("Frekans")
plt.show()
```

**Yorum:** Sipariş tutarlarının büyük çoğunluğu 0-100£ aralığında yoğunlaşmaktadır. 
Luvionis bu dağılımı kullanarak düşük/yüksek değerli müşteri segmentlerini ayırt eder.

```python
top5_countries = df[df["Country"] != "United Kingdom"]["Country"].value_counts().head(5).index
df_top5 = df[df["Country"].isin(top5_countries)]
sns.boxplot(data=df_top5, x="Country", y="TotalPrice")
plt.title("Ülkeye Göre Sipariş Tutarı Dağılımı (UK hariç, Top 5)")
plt.ylabel("Sipariş Tutarı (£)")
plt.xticks(rotation=20)
plt.show()
```

**Yorum:** Hollanda ve Almanya'da yüksek değerli aykırı siparişler dikkat çekmektedir. 
Luvionis bu müşteri segmentine özel fiyatlama stratejisi önerir.

```python
pivot = df.groupby(["DayOfWeek", "Month"])["TotalPrice"].sum().unstack()
pivot.index = ["Pzt","Sal","Çar","Per","Cum","Cmt","Paz"]
sns.heatmap(pivot, cmap="YlOrRd", fmt=".0f", annot=False)
plt.title("Gün × Ay Bazında Toplam Ciro Isı Haritası")
plt.xlabel("Ay")
plt.ylabel("Gün")
plt.show()
```

**Yorum:** Kasım ayı ve hafta içi günleri en yüksek ciroya ulaşılan dönemlerdir. 
Luvionis, reklam bütçelerinin bu dönemlerde artırılmasını önerir.

```python
df[df["Country"] != "United Kingdom"].groupby("Country")["TotalPrice"].sum().sort_values(ascending=False).head(8).plot(kind="bar", color="coral", edgecolor="white")
plt.title("Ülke Bazında Toplam Ciro (UK hariç, Top 8)")
plt.ylabel("Toplam Ciro (£)")
plt.xticks(rotation=30)
plt.show()
```

**Yorum:** Hollanda ve EIRE (İrlanda) UK dışında en yüksek ciroyu üreten pazarlardır.

```python
monthly = df.groupby("Month")["TotalPrice"].sum().reset_index()
monthly.columns = ["Ay", "Toplam Ciro"]
fig = px.bar(monthly, x="Ay", y="Toplam Ciro",
             title="Aylık Toplam Ciro — Luvionis Trend Analizi",
             color="Toplam Ciro", color_continuous_scale="Blues")
fig.show()
```

**Yorum:** Kasım ayında belirgin bir ciro zirvesi görülmektedir (Black Friday etkisi). 
Luvionis bu trendi nedensel modeliyle doğrulayarak reklam önerileri sunar.

```python
country_summary = df[df["Country"] != "United Kingdom"].groupby("Country").agg(
    Ciro=("TotalPrice","sum"),
    SiparisAdet=("Invoice","nunique"),
    OrtSiparis=("TotalPrice","mean")
).reset_index().sort_values("Ciro", ascending=False).head(15)

fig = px.scatter(country_summary, x="SiparisAdet", y="Ciro",
                 size="OrtSiparis", color="Country", hover_data=["OrtSiparis"],
                 title="Ülke Bazında Sipariş Adedi vs Ciro (Luvionis Segment Haritası)")
fig.show()
```

**Yorum:** Bu scatter plot, yüksek sipariş adedi ama düşük ortalama sepet değeri olan 
ülkeleri Luvionis'in upsell hedefleri olarak işaretlemesine olanak tanır.

## 4. Bulgular ve İçgörüler

1. **Mevsimsellik:** Kasım ayı ciroda belirgin zirve oluşturmaktadır; Luvionis reklam 
   bütçelerinin Ekim sonundan itibaren artırılmasını önerir.
2. **Coğrafi fırsat:** Hollanda ve İrlanda, yüksek ortalama sepet değeriyle premium 
   segment potansiyeli taşımaktadır.
3. **Sepet yoğunluğu:** Siparişlerin büyük çoğunluğu 0-100£ aralığındadır; ortalama 
   sepet değerini artıracak cross-sell stratejileri doğrudan ciroya yansıyacaktır.
4. **Hafta içi dominansı:** Ciro hafta içinde yoğunlaşmaktadır; B2B müşteri 
   profilinin ağırlıklı olduğunu göstermektedir.
